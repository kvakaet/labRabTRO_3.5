построить круговую диаграмму по общему объему продаж за 5 лет

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = np.array([2015, 2016, 2017, 2018, 2019])
product1_sales = np.array([100, 120, 80, 90, 110])
product2_sales = np.array([80, 90, 100, 110, 120])
product3_sales = np.array([110, 100, 120, 80, 90])

total_sales = product1_sales + product2_sales + product3_sales

plt.pie(total_sales, labels=years, autopct='%1.1f%%', startangle=90)

plt.title('Общий объем продаж по годам')

plt.show()
```
