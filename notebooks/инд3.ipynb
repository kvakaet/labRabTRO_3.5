построить линейный график годового прироста ввп.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Годы и соответствующие значения ВВП

```python
years = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019])
gdp_values = np.array([14.48, 15.54, 16.2, 16.69, 17.42, 17.95, 18.52, 19.48, 20.51, 21.43])
```

Вычисление годового прироста ВВП

```python
gdp_growth = np.diff(gdp_values) / gdp_values[:-1] * 100
```

Построение линейного графика годового прироста ВВП

```python
plt.plot(years[1:], gdp_growth, marker='o', linestyle='-', color='blue')
plt.xlabel('Год')
plt.ylabel('Годовой прирост ВВП (%)')
plt.title('Годовой прирост ВВП')
plt.grid(True)
plt.show()
```
