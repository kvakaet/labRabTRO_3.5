# Идивидуальное задание №1

 Сгенерировать синусоиду с шумом, применить линейную регрессию для подгонки линии к данным и построить график данных и линейной регрессии с использованием библиотеки matplotlib.pyplot. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def compute_sine_wave(amplitude, frequency, phase, num_points):
    x = np.linspace(0, 2*np.pi, num_points)
    y = amplitude * np.sin(frequency * x + phase)
    return x, y

def compute_linear_regression(x, y):
    n = len(x)
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    xy_mean = np.mean(x * y)
    x_squared_mean = np.mean(x**2)

    slope = (xy_mean - x_mean * y_mean) / (x_squared_mean - x_mean**2)
    intercept = y_mean - slope * x_mean

    return slope, intercept

def plot_regression(x, y, slope, intercept):
    plt.scatter(x, y, color='blue', label='шум')
    regression_line = slope * x + intercept
    plt.plot(x, regression_line, color='red', label='линейная регрессия')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Линейная Регрессия')
    plt.legend()
    plt.show()
```

Генерируем синусоиду с шумом

```python
amplitude = 2.5
frequency = 1.5
phase = np.pi/4
num_points = 100
noise = np.random.normal(0, 0.5, num_points)
x, y = compute_sine_wave(amplitude, frequency, phase, num_points)
y_noisy = y + noise
```

Применяем функцию линейной регрессии и строим график данных и линейной регрессии

```python
slope, intercept = compute_linear_regression(x, y_noisy)

plot_regression(x, y_noisy, slope, intercept)
```
