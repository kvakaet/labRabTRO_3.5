построить сболбчатую диаграмму по продажам 3 продуктов за 5 лет

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = np.array([2015, 2016, 2017, 2018, 2019])
product1_sales = np.array([100, 120, 80, 90, 110])
product2_sales = np.array([80, 90, 100, 110, 120])
product3_sales = np.array([110, 100, 120, 80, 90])

bar_width = 0.25

r1 = np.arange(len(years))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, product1_sales, color='blue', width=bar_width, edgecolor='black', label='Апельсины')
plt.bar(r2, product2_sales, color='orange', width=bar_width, edgecolor='black', label='Мандарины')
plt.bar(r3, product3_sales, color='green', width=bar_width, edgecolor='black', label='Бананы')

plt.xlabel('Годы')
plt.ylabel('Объем продаж')
plt.title('Продажи продуктов по годам')
plt.xticks([r + bar_width for r in range(len(years))], years)

plt.legend()

plt.show()
```
