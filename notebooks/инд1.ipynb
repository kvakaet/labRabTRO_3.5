# индивидуальное задание №1

Изучить динамику цен акций двух компаний, построить линейный график сравнения и выявить возможную корреляцию между двумя временными рядами

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
ax = plt.gca()
prices_A = [5, 8, 15, 13, 14]
prices_B = [7, 4, 1, 2, 1]

ax.plot(prices_A, marker='o', linestyle='-', color='b', label='Company A')
ax.plot(prices_B, marker='o', linestyle='-', color='r', label='Company B')

ax.set_xlabel('Время')
ax.set_ylabel('Цена акции')
ax.set_title('Динамика цен акций компаний A и B')
ax.legend()

plt.show()
correlation = np.corrcoef(prices_A, prices_B)[0, 1]
print(f"Коэффициент корреляции: {correlation}")
```
